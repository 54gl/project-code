```python
# Import packages.
import cvxpy as cp
import numpy as np
from typing import Any, Callable, Dict, Optional, Tuple
from numpy import array, ndarray
```

```python
def cvxpy_solve_qp(n,P, q, G, h, A, b, solver: str = None,
                    initvals: Optional[ndarray] = None,
                   ) -> Optional[ndarray]:
    
    # Define and solve the CVXPY problem.
    x = cp.Variable(n)
    prob = cp.Problem(cp.Minimize((1/2)*cp.quad_form(x, P) + q.T @ x),[G @ x <= h,A @ x == b])
    prob.solve()
 #   return (prob.solve())
  #  return array(x).reshape((q.shape[0],))
    return(x.value)
```

```python
# Import packages.
import cvxpy as cp
import numpy as np
m = 5
n = 70
p = 5
P = np.loadtxt('P.txt')
q = np.loadtxt('q.txt')
G = np.loadtxt('G.txt')
h = np.loadtxt('h.txt')
A = np.loadtxt('A.txt')
b = np.loadtxt('b.txt')

x=cvxpy_solve_qp(n, P,q, G,h,A,b)
v=0.5*x.T@P@x+q.T@x
print(v)
```
